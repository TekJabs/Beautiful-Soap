Web Scraping using **Beautiful** **Soup**

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd

#Adding the URl of the website to be scrapped
URL = "https://hgrill.netlify.app/"

#Making HTML variable
html_text = urllib.request.urlopen(URL)

#Making a soup object
soup = BeautifulSoup(html_text,"html.parser")

#Second Page
class burger_section:


  def burger_names(self):
    burger_page = soup.find('div' , id="menu-page")
    burger_names = burger_page.find_all('p',class_="burger-name")
    #List of burger names
    burger_names_text = []

    print("The options are as follows :- ")
    print("Press 1 to display all burgers")
    print("Press 2 to filter burgers by condition")

    #Asking the user to enter the option


    option = input("Please enter you desired option:  ")
    if(option=='1'):
         with open('burger_wesbite.txt','a') as f:
          f.write("The burgers are as follows :- \n\n")
          for i in range(0,len(burger_names)):
              burger_names_text.append(burger_names[i].text)
              f.write(f"{burger_names[i].text} \n")

    #Making a data Frame of burgers for Pandas

    data = {
        "Burgers":burger_names_text
    }

    #Writing on a CSV file
    df = pd.DataFrame(data)
    df.to_csv("burger_website.csv")


    #This section was for learning purpose
    if(option=='2'):
      filter_price = input("Enter the price: ")
      for i in range(0,len(burger_names)):
        if filter_price in burger_names[i].text:
          print(burger_names[i].text)



  def __init__(self):
    self.burger_names()






#First page
class main_section:

  #main_section Nav Funtion

  def nav_data(self):
    navbar = soup.find('nav')
    #scrapping the LI's of the navbar

    #list container is an array and can be used with indexes
    list_container = navbar.find_all('li')

    list_li_text = []

    for i in range(0,len(list_container)):
      list_li_text.append(list_container[i].text)


    #Writing the nav data in the file
    with open('burger_wesbite.txt','w') as f:
        f.write("This is the nav data \n")
        for i in range(0,len(list_container)):
          f.write(f"{list_li_text[i]} \n")


  #main_section banner text

  def banner_text(self):

    #banner is the picture section of the website
    banner = soup.find('div', class_="home-content")
    #Banner p's for the text of the banner
    banner_ps = banner.find_all('p')
    with open('burger_wesbite.txt','a') as f:
      f.write("Data of the banner \n")
      for i in range(0,2):
        #we added this if condition beacuase we needed insexes 0 and 2
        if(i==1):
          i = i + 1
        f.write(f"{banner_ps[i].text} \n")


  #This is the contrcutor
  def __init__(self):
    self.nav_data()
    self.banner_text()


#Function holding all object calls
def main():
  obj1 = main_section()
  obj2 = burger_section()

main()
```
